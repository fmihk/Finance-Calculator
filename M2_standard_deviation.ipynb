```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the CSV file
df = pd.read_csv('/content/M2SL.csv')
df['DATE'] = pd.to_datetime(df['DATE'])

# Filter the data for dates starting from
df = df[df['DATE'] >= '2020-01-01']

# Calculate the mean and standard deviation of M2SL for one year
one_year_ago = df['DATE'].max() - pd.DateOffset(years=1)
one_year_data = df[df['DATE'] >= one_year_ago]['M2SL']
m2sl_mean = np.mean(one_year_data)
m2sl_std = np.std(one_year_data)

# Calculate the upper and lower bounds based on one standard deviation away from the mean
m2sl_upper_bound = m2sl_mean + m2sl_std
m2sl_lower_bound = m2sl_mean - m2sl_std

# Plot the data and bounds on a line plot
plt.plot(df['DATE'], df['M2SL'], label="M2SL")
plt.axhline(y=m2sl_mean, color='r', linestyle='--', label="Mean")
plt.axhline(y=m2sl_upper_bound, color='g', linestyle='--', label="Upper Bound")
plt.axhline(y=m2sl_lower_bound, color='g', linestyle='--', label="Lower Bound")
plt.legend()
plt.show()
```
