```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

end_date = datetime.today().strftime('%Y-%m-%d')
n225 = yf.download("^N225", start="2020-01-01", end=end_date)
usdjpy = yf.download("USDJPY=X", start="2020-01-01", end=end_date)
data = pd.merge(n225, usdjpy, on='Date')
data['adj_n225'] = data['Close_x'] * data['Close_y']

# plot the adjusted Nikkei
plt.plot(data.index, data['adj_n225'])
plt.title('Nikkei 225 adjusted by USD/JPY')
plt.xlabel('Date')
plt.ylabel('Index value')
plt.show()
```
